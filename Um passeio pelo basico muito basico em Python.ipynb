É possivel adicionar elementos em um set usando os métodos .add() e .update:

`````python
# .add()
my_set = set([1,2,2,2,3,3,3,4,4,4,5,5,7,7,8,9,9,9])
my_set.add(10)
my_set
`````

Uma propriedade interessande de sets, é que eles suportam operações matemáticas de conjunto, tais como união, intersecção e diferença simétrica, verificação se os conjuntos são disjuntos ou subconjutos, como os exemplos a seguir. 

Em geral realizar essas operações é mais facil e eficiente com sets do que com listas ou tuplas

`````python
#União
set1 = {'A','B','C'}
set2 = {'B','C','D'}

set3=set1.union(set2)
set3
`````

`````python
#Interseção 
set1 = {'A','B','C'}
set2 = {'B','C','D'}

set3=set1.intersection(set2)
set3
`````

`````python
#Diferença simétrica, tudo que está em set1 mas nao em set2, 
#e tudo que está em set2 mas não em set1

set1 = {'A','B','C'}
set2 = {'B','C','D'}

set3=set1.symmetric_difference(set2)
set3
`````

`````python
#Verificação se é um subconjunto:

set1 = {'B','C'}
set2 = {'B','C','D'}

set3=set1.issubset(set2)
set3
`````

`````python
#Verificação se são disjuntos, ou seja, nao tem elementos em comum: 

set1 = {'A','B','C'}
set2 = {'D','E','F'}

set3=set1.isdisjoint(set2)
set3
`````

Como sets são não ordenados, eles não suportam slicing com a notação de colchetes '[ ]' como listas e tuplas suportam:

`````python
set1 = {'A','B','C'}
set1[0]
`````

## 8 - Funções

Dentro desses parêntesis são passados os *parâmetros* que a função ira precisar. Para definir a lógica interna da função, o código deverar estar identado dentro dela.

Vamos definir uma função simples que some dois números:

A lógica interna de uma função não precisa ser simplória como a do primeiro exemplo. Podemos pensar códigos mais complexos. Por exemplo, podemos incrementar a logica do primeiro exemplo para que seja retornado o valor da soma de uma lista de valores: 

`````python
def soma_lista(valores):
  'soma uma lista de valores'  
  total = 0
  for valor in valores: #Não se preocupem com esses loops, falermos sobre isso mais pra frente
    total = total + valor   
  return total 
`````

`````python
def soma_lista(valores): 
  'soma uma lsta de valores'
  total = 0 
  for valor in valores:
    total = total + valor
  return total
`````

`````python
soma_lista([1,2,3,4,5])
`````

Funções podem retornar mais de um valor. Os resultados serão retornados em uma tupla ordenada de acordo com o que foi definido no return da função:

`````python
def soma_e_len_lista(valores):
  '''Calcula a soma e o tamanho
  de uma lista de valores'''  
  total = 0
  for valor in valores:   
    total = total + valor 
  
  tamanho = len(valores)

  return total, tamanho
`````

`````python
soma_e_len_lista([1,2,3,4])
`````

## 9 - Funções anônimas ou *lambda*

Em alguns contextos, é comum querermos criar funções de maneira rápida para serem usadas de maneira concisa em algum ponto do código. 

Para esses casos existe a palavra reservada 'lambda' e uma sintaxe que permite criar uma função de maneira mais concisa do que uma declaração convencional: 

`````python
# uma função anônima para calcular o quadrado de números
a = lambda x: x**2 
a(2)
`````

Em análise de dados, frequentemente funções que transformam os dados aceitam como argumento outras funções, nesse caso, funções lambda podem ser uma opção sucinta para se definir a transformação

Um bom exemplo disso é a função map(), que aplica uma transformação definida por uma função a uma coleção python.

Por exemplo, se quisermos obter a raiz quadrada de todos os elementos de uma lista, poderiamos fazer:

`````python
def raiz(x):
  result= x**0.5
  return result

my_list=[1,2,3,4]
raizes = list(map(raiz,my_list))

raizes
`````

O trecho de código acima produz o mesmo resultado que o seguinte:

`````python
raizes = list(map(lambda x: x**0.5,[1,2,3,4]))
raizes
`````

Em geral funções lambdas funcionam bem para definições de funções simples, se uma maior complexidade for exigida, pode ser preferível usar uma definição convencional de função.

### If, Else 

Um bloco if/else verifica uma condição para executar o trecho de código identado subsequente. Caso a condição nao seja avaliada como True, o trecho identado subsqunte ao Else é executado:

### If, Elif, Else

Um bloco If/Elif/Else funciona de maneira similar, contudo a condição do Elif também é avaliada. Pode se pensar num bloco if/Elif/Else como um bloco If/Else de mais de uma condição 

`````python
a = 7
b = 7 

if a<b:
  print(f'{a} é menor que {b}')
elif a == b:
  print(f'{a} é igual a {b}')
else: 
  print(f'{a} é maior que {b}')
`````

### Loop For

Um loop (ou laço) for é uma maneira de se executar uma série de instruções de maneira subsequente sem a necessidade de se escrever explicitamente cada uma delas. 

Vamos usar um exemplo simples, suponha que a gente tenha uma lista de nomes e queremos printar a primeira letra de cada um deles:

`````python
#poderíamos fazer:
cavaleiros = ['Saga','Shaka','Aiolia','Mu']

print(cavaleiros[0][0])
print(cavaleiros[1][0])
print(cavaleiros[2][0])
print(cavaleiros[3][0])
`````

Facilmente se percebe que isso pode se tornar uma tarefa maçantemente insuportável caso a nossa coleção de nomes fosse maior. 

Nesse tipo de situação usamos um loop for, que em python tem uma sintaxe bastante elegante:

`````python
cavaleiros = ['Saga','Shaka','Aiolia','Mu']

for nome in cavaleiros:
  print(nome[0])
`````

`````python
cavaleiros = ['Saga','Shaka','Aiolia','Mu']

for cavaleiro in cavaleiros:
  print(cavaleiro)
`````

A serie instruções dentro do loop pode ser mais complexa, e em geral será. Mas a estrutura será a mesma.

A sintaxe acima funciona nao apenas para listas, mas para todo objeto python *iterável*. Todas as coleçoes de dados vistas anteriormente são estruturas iteráveis, bem como outros objetos comuns em análise de dados, como arrays numpy, Pandas Series, colunas de DataFrames.

Lists Comprehension 

Muito frequentemente você se verá usando trechos de código com o seguinte formato:

``` python
lista_vazia = []
for um_trem in colecao_de_trens:
  resultado = faz_algo_com_o_trem(trem)
  lista_vazia.append(resultado)  
````
Em que o resultado de cada iteração do loop é 'appendado' ao final de uma lista criada inicialmente vazia. O resultado disso é uma lista contendo todos os resultados de uma transformação sobre os elementos de um iterável.

Existe em Python uma alternativa ao bloco de código acima, chamada *List Comprehension* que de maneira mais concisa realiza a mesma tarefa. Ela assume forma geral a seguir:

``` python
[faz_algo for trem in colecao_de_trens]
```

Também é frequênte *Lists Comprehensions* para se verificar alguma condição antes de se tranformar o elemento do iterável.

``` python
[faz_algo for trem in colecao_de_trens if condicao]
```

Vamos retomar o exemplo com os nomes de alguns Cavaleiros de Ouro:


`````python
cavaleiros = ['Saga','Shaka','Aiolia','Mu']

iniciais = [] #Isso aqui é muito comum, sério
for nome in cavaleiros:
  inic = nome[0]
  iniciais.append(inic)

iniciais 
`````

O trecho acima pode ser reescrito apenas como:

`````python
[cavaleiro[0] for cavaleiro in cavaleiros]
`````

Se quiséssemos testar alguma condição, por exemplo cavaleiros cujos nomes não comecem com 'S', ficaria:

`````python
[cavaleiro[0] for cavaleiro in cavaleiros if cavaleiro[0] != 'S']
`````

Lists comprehensions, por sua forma concisa, são muito práticas. 
Contudo, acredito que funcionem melhor com transformações simples e checagem de condições também simples. Lists Comprehentions que utilizam lógicas complicadas são menos legíveis do que o loop for 'por extenso'.

### 11.2 - Dados os seguintes filmes:
- Super Xuxa Contra o Baixo Astral
- Vingadores Ultimato
- Homem Aranha Longe de Casa
- Batmam Begins
- O Poderoso Chefão
- Forest Gump
- Star Wars o Imperio Contra Ataca
- O Auto da Compadecida

Vamos criar um programinha que:
- 1) Sorteie 3 filmes da lista
- 2) Peça ao usuário uma nota de 0 a 5 sobre esse filme
- 3) Apresente os filmes junto com suas notas
- 4) Calcule a nota média dada e apresente esse resultado

Dica, a função random.sample() realiza sorteios sem reposição a partir de uma coleção. 

`````python
### vamos primeiro criar uma lista com os filmes:
filmes = ["Super Xuxa Contra o Baixo Astral",
          "Vingadores Ultimato",
          "Homem Aranha Longe de Casa",
          "Batmam Begins",
          "O Poderoso Chefão",
          "Forest Gump",
          "Star Wars o Imperio Contra Ataca",
          "O Auto da Compadecida"]
`````

`````python
#Agora vamos usar a função random.sample() para sortear os filmes: 
import random
filmes_escolhidos = random.sample(filmes,k=3) #realizando um sorteio de 3 filmes sem repetição
`````

`````python
#Agora vamos criar um loop que peça ao usuário uma nota para cada um dos filmes
#sorteados, checando se a nota fornecida é valida 

def check_range(n, upper, lower):
  '''checa se a nota n está entre os limites
  upper e lower'''
  upper_check = n<=upper
  lower_check = n>=lower  
  return upper_check and lower_check

notas = []
for titulo in filmes_escolhidos:
  print(f'Por favor dê uma nota de 0 a 5 ao filme {titulo}:')
  nota = int(input())
  valid = check_range(nota,5,0)

  while valid == False:
    print(f'Por favor dê uma nota de 0 a 5 ao filme {titulo}:')
    nota = int(input())
    valid = check_range(nota,5,0)  
  notas.append(nota)

#agora vamos salvar as notas dadas aos filmes escolhidos
#em um dicionário
filmes_notas_dict =dict(zip(filmes_escolhidos,notas))

#calculando a média das notas
nota_media = sum(notas)/len(notas)

#Agora vamos compor strings para apresentar os resultados:
sub_strings = []
for titulo in filmes_notas_dict.keys():
  s=f'A nota dada ao filme {titulo} foi: {filmes_notas_dict[titulo]}'
  sub_strings.append(s)

for string in sub_strings:
  print(string)
print(f'A nota média dada foi: {nota_media}')
`````
